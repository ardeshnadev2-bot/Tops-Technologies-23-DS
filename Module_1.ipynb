Write a Python program to count the occurrences of each word in a given sentence.

```python
sentence = input("Enter sentence: ").split()
freq = {}
for word in sentence:
    freq[word] = freq.get(word, 0) + 1
print(freq)
```

Write a Python program to find the first appearance of the substring 'not' and'poor' from a given string, if 'not' follows the 'poor', replace the whole 'not'...'poor' substring with 'good'. Return the resulting string

```python
s = input("Enter string: ")

not_pos = s.find("not")
poor_pos = s.find("poor")

if not_pos != -1 and poor_pos != -1 and not_pos < poor_pos:
    s = s.replace(s[not_pos:poor_pos+4], "good")

print(s)
```

Write a Python program to check whether a list contains a sublist.

```python
def contains_sublist(lst, sub):
    for i in range(len(lst) - len(sub) + 1):
        if lst[i:i+len(sub)] == sub:
            return True
    return False
```

Given a number n, write a python program to make and print the list of Fibonacci series up to n. Input : n=7 Hint : first 7 numbers in the series Expected output :
First few Fibonacci numbers are 0, 1, 1, 2, 3, 5, 8, 13

```python
n = int(input("Enter n: "))
fb = [0, 1]

for i in range(2, n):
    fb.append(fb[-1] + fb[-2])

print(fb)
```

Write a python program using function to find the sum of odd series andevenseries
Odd series: 12/ 1! +32/ 3! + 52/ 5!+……n
Even series: 22/ 2! + 42/ 4! + 62/ 6!+……n

```python
import math

def odd_series(n):
    s = 0
    for i in range(1, n+1, 2):
        s += (i**2) / math.factorial(i)
    return s

def even_series(n):
    s = 0
    for i in range(2, n+1, 2):
        s += (i**2) / math.factorial(i)
    return s
```

 Python Program to Find Factorial of Number Using Recursion

```python
def fact(n):
    if n == 0:
        return 1
    return n * fact(n - 1)

print(fact(int(input())))
```
